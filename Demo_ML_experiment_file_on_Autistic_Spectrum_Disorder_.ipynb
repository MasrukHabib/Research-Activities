#Loade the dataset and check the data

```python
import pandas as pd
df = pd.read_csv("/content/Autism_Child_Data.csv")
df.head()
```

# Remove the leading "b'" and the trailing " ' "

```python
df= df.replace(r"^b'|'$", '', regex=True)
df.head()
```

# Replace categorical values with numerical representations for gender and yes/no features

```python
df.replace({'m':1.0,'f':0.0,'yes':1.0,'no':0.0,'NO':0.0,"YES":1.0}, inplace=True)
df.head()
```

# Create a list of unknown characters to be replaced with NaN

```python
unknown_chars = ['?']
```

```python
import numpy as np
df.replace(unknown_chars, np.nan, inplace=True)
```

```python
df.isnull().sum()
```

# Fill missing values using forward fill.


**You can also use these methods to handle missing values:**

**df.bfill():** This uses backward fill, propagating the next known non-missing value backward to fill gaps.

**df.fillna(method='ffill'):** This is another way to perform forward fill.

**df.fillna(method='bfill'):** This is another way to perform backward fill.

**df.fillna(value):** This replaces missing values with a specified value. For example, df.fillna(0) fills missing values with 0.

**df.dropna():** This removes rows or columns containing missing values. You can use the axis parameter to specify whether to drop rows (0) or columns (1).

```python
df=df.ffill()
df.isnull().sum()
```

```python
# Define feature matrix X and target vector y
X = df.drop('Class/ASD', axis=1)
y = df['Class/ASD']
```

# Split data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
import pandas as pd

# Calculate class distribution
class_distribution = df['Class/ASD'].value_counts()

# Print class distribution
print("Class Distribution:")
print(class_distribution)

# Check if data is balanced
balanced = class_distribution.min() / class_distribution.max() >= 0.5

if balanced:
    print("The data is balanced.")
else:
    print("The data is imbalanced.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plotting the balance of the target variable
plt.figure(figsize=(10, 5))

# Bar plot
plt.subplot(1, 2, 1)
sns.countplot(x=y)
plt.title('Class Distribution')
plt.xlabel('Class/ASD')
plt.ylabel('Count')

# Pie chart
plt.subplot(1, 2, 2)
y.value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Class Distribution')
plt.ylabel('')  # Hide y-label for the pie chart

plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize and train the classifier
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = classifier.predict(X_test)

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu")

# Add labels, title and show the plot
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

#Model Traning

#Using NB

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Naive Bayes model
nb_model = GaussianNB()

# Train the model
nb_model.fit(X_train, y_train)

# Make predictions
y_pred = nb_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(report)
```

#Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors

# Train the model
knn_model.fit(X_train, y_train)

# Make predictions
y_pred = knn_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(report)
```

#Using DT

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)

# Make predictions
y_pred = dt_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(report)
```
